# Numerical Data Type

```python
# Python Data type- 

#1. int - To store integer value
#You can store positive as well as negative integer numbers.
#You can store integer value other than base 10 such as binary (base 2), octal(base 8), hexadecimal (base 16).

n = 10
print(n)
print(type(n))



print("-------------------------------------------")



#2. float - To store floating -point values or decimal value 

#Floating point values can be represented using the exponential form , also called scintific notations.

#The benifit of using the exponential form to represent floating-point values is we can 
#represent large values using less memory.

n = 10.2
print(n)
print(type(n))

#Exponential float number

num1 = 1.22e4 # 1.22e4 = 1.22 X 10 raise to power 4 = 12200

print(num1)
print(type(num1))



print("-------------------------------------------")



#3.complex - To store complex number (real and imaginary part)

# It is number with real and imaginary compponent represented as a + bj
#where a and b contain integer or floating values.

#Real part of complex no. can be int, float, binary
#But the imaginary part should be represented using decimal form only.

#If we are trying to represent an imaginary part as binary, hexadecimal or octal it will show error.

x = 5 + 2j # Both are int type 
print(x)
print(type(x))

y = 10 + 1.2j #one int and one is float
print(y)

z = 10.2 + 5.2j #Both val are float
print(z)



```


# Sequence Data Type

```python
'''
1. str data type - String is a sequence of character enclosed within single or double quote.

This character could be anything like letters, numbers or special symbols enclosed within single or
double quotation marks.

Once a string is created, we can do many operations on it, such as searching substring inside it,
splitting string etc.

The string is immutable i.e. it can not be changed once it defined.

Non-changable behaviour is called immutability.

'''
name = "Jessa"

print(name)
print(type(name))

# 0  1  2  3  4  --> Index value
# j  e  s  s  a  --> Actual string

#print(name[0]) = "K" # --> cannot modify string when it once created bcoz strings are immuatable.

print('Accessing a letter from a given string ',name[3])

print("------------------------------------")

'''
List Data Type - Ordered Collection of elements (also known as sequence)

List element can be accessed, iterated and removed accoreding to the order they inserted at the time
of creation.

The list data type represents groups of element as a single entity.

The list has following properties :- 

1.Mutable - The elements of the list can be modified.
          - We can add or remove items to the list after it has been created.
          
2.Orderd  - The items in the list are ordered.
          - Each item has a unique index value.
          - The new items will be added to the end of the list.
          
3.Heterogenous - The list can contain different kinds of element.
               - i.e. int,str,bool,list,tuple etc. any type.

4.Duplicates   - The list can be contains duplicates.
               - i.e. they can contain two items with same values.

list can be created using two types -

1. list() constructor  
2. using square brackets [] 

'''

#Example

#1.list() constructor

my_list = list((1,2,3))
print(my_list)

print("-------------------------------------")

for i in [10,20,"jessa",[10,20,"Roshni"],25.8]:
    print(i)
    print(type(i))

    
print("------------------------------------")

#empty list

my_list = []

print(my_list)


print("-------------------------------------")

'''
Tuple data type - Ordered Collection of elements that are unchangeable (immutable).

The tuple is same as the list, except the tuple is immutable.
Tuple is read-only version of list.

For Example - 

If you want to store the roll numbers of the students that you don't change ,you can use the tuple to 
store it.

Note: Tuple maintains the insertion order and also, allows us to store duplicate elements.

we can create tuple using two ways : 

1.tuple()
2.using parenthesis ()

'''

my_tuple = (11,24,56,88,78)
print(my_tuple)
print(type(my_tuple))

print(my_tuple[2])

# my_tuple[2] = 12 - cannot change value because tuple is immutable
```

# Dictionary

```python
'''
Dictionary Data type - Unordered collection of values stored in the form of key-value pairs.

EX - If you want to store the name and roll number for all students, then you can use the dict. 
data type

In Dictionary , Duplicate keys are not allowed but duplicate values can be inserted.

So if we try to insert duplicate value with duplicate key , the old value will be repalced with new 
value.

Dictionary has some charcteristics:- 
1. Heterogeneous : Values of all data types are allowed.
2. Dictionary is mutable which means we can modify its items.
3. Dictionary is unordered hence we cannot perform indexing and slicing on it.

We can create dict using two ways
1. Using curly brackets {}
2. dict({...})



'''
# {}
my_dict = {1 : "Smith", 2: "Emma", 3: "Jessa"}
print(my_dict)
print(type(my_dict))

#dict()

hadoop ="elephant toy"
java ="coffee"
python = "Monty python"

dict1 = dict({"python": "Monty python", "java" : "coffee", "hadoop" : "elephant toy"})
dict2 = dict({1: "Monty python", 2 : "coffee", 3 : "elephant toy"})
print(dict1)
print(dict2)
```

# Set

```python
'''
Set Data Type - Unordered collection of data item ithat are unique.
              - Duplicate elements are not allowed 
              - Heterogeneous (values of all data type) elements are allowed.
              - insertion order is not preserved(unordered), so we cant perform indexing and slicing.
              - Set used to represent unique group of values as single entity.
              
We can create set using two ways ;

- By enclosing elements in the curly brackets {}
- Using a set() method
              
'''
#using {}
my_set = {100,20,25,25}
print(my_set)
print(type(my_set))

#set({...})
my_set1 = set({100,25.75,"jessa"})
print(my_set1)
print(type(my_set1))

print("------------------------------------------")

'''
FrozenSet - Create immutable set

Note: If we try to perform operations like add, remove on frozen set , will receive an error.

use frozenset() class for creating frozenset.

'''

f_set = frozenset(my_set)
print(f_set)
print(type(f_set))

#f_set.add(300) --> frozenset is immutable
```
